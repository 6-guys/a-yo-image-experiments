```python
from PIL import Image
import numpy as np

dir = '/content/drive/MyDrive/NIPA/a-yo-image/initial-experiments/mongoose/separator_output/output_frames/'
data_stack = []
for j in range(50) :
  img_stack = []
  for i in range(10) :
  # 이미지 파일 불러오기
    img = Image.open(dir + 'frame_'+ str(j) +'_' + str(i) + '.png')

    # 이미지를 (h, w, c) 형식의 NumPy 배열로 변환
    img_stack.append(img)
  img_stack = np.stack(img_stack)
  data_stack.append(img_stack)
  # 배열의 모양 출력 (옵션)
  print("Image array shape:", img_stack.shape)

data_stack = np.stack(data_stack)
print("Image array shape:", data_stack.shape)
```

```python
np.save('data_run.npy', data_stack)
```
